```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
path='/content/car_prices.csv'
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
year_wise_sales=df['year'].value_counts()
year_wise_sales
```

```python
plt.bar(year_wise_sales.index,year_wise_sales.values)
plt.title('Year wise sales report',color='maroon')
plt.xlabel('year')
plt.ylabel('sales')
```

The above report indicates that year '**2012**' saw the highest number of sales with 102315 units being sold that year followed by 2013 at 98168 units and then 2014 at  81070 units

```python
top_brand_wise=df['make'].value_counts().head(10)
top_brand_wise
```

```python
plt.style.use('dark_background')
plt.bar(top_brand_wise.index,top_brand_wise.values,color='red')
plt.title('Brand Wise Sales Report')
plt.xlabel('Brand')
plt.ylabel('Sale')
plt.xticks(rotation=90)
```

The above generated report indicates that car manufacturer FORD generated the highest sales with a whopping 93554 units sold followed by CHEVROLET at 60197 unit sales and NISSAN at 53946 unit sales. This indicates the dominance of American manufacturers in the American soil

```python
transmission_wise=df['transmission'].value_counts()
transmission_wise
```

```python
my_explode=[0.1,0,0,0]
plt.title('market cap by by transmission type',color='maroon')
colors=['red','green','orange','yellow']
plt.pie(transmission_wise,labels=transmission_wise.index,explode=my_explode,autopct='%1.1f%%',colors=colors)
```

The above reort indicates the market presence of transmission type used in the vehicels. The report shows that the automatic transmission dominates the market with 96.4% of the vehicles being equippped with an Automatic transmission   

```python
plt.scatter(df['odometer'],df['mmr'],color='red',marker='*')
plt.title('Correlation between mileage and mmr')
plt.xlabel('Mileage')
plt.ylabel('price')
```

The above scatter plot analysis illustrates as the vehicle mileage increases the market price for the vehicle decreased. The market price is inversly proportional to vehicle mileage.

```python
kia_filter= df[df['make']=='kia']
kia_filter
```

```python
kia_filter['model'].unique()
```

```python
variant=kia_filter.groupby('model')['sellingprice'].mean()
variant
```

```python
plt.bar(variant.index,variant.values,color='yellow',edgecolor='red')
plt.title('Average selling price report for kia models',color='maroon')
plt.xlabel('Model')
plt.ylabel('Price')
```

 The above barplot illustrates the average selling price for Kia variants with the model **Borrego** having the highest average selling price followed by Rio and Sportage and then Optima.

```python
top_market_price_by_model=df.groupby('model')['mmr'].max().sort_values(ascending=False).head(10).reset_index()
top_market_price_by_model
```

```python
plt.bar(top_market_price_by_model['model'],top_market_price_by_model['mmr'],color='yellow',edgecolor='green')
plt.title('Top 10 model by market price')
plt.xlabel('Model')
plt.ylabel('Market Price')
plt.xticks(rotation=90)
```

The above barplot analysis indicates that the Ferrari's car model 458 Italia has the highest market value among all the models at 182000 USD followed by RollsRoyce Ghost at 178000 and BMW i8 at 176000

```python
top_market_price_by_brand=df.groupby('make')['mmr'].sum().sort_values(ascending=False).head(10).reset_index()
top_market_price_by_brand
```

```python
plt.bar(top_market_price_by_brand['make'],top_market_price_by_brand['mmr'])
plt.title('Top brands with highest market value')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=90)
```

The above barplot analysis demonstrates the brands with top market value with all models combined. The graph is topped by the American manufaturer FORD at 1378388875  followed by Chevrolet at 731431500. This clearly demonstrates the dominance and market demand for the American manufaturers.
